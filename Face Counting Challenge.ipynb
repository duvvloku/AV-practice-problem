```
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from tqdm import tqdm
```

```
test = pd.read_csv('/content/drive/My Drive/HAX/face_count/test_Rj9YEaI.csv')
print(test.head())
```

```
faces=[]
facedetector=cv2.CascadeClassifier("/content/Face-detection/haarcascade_frontalface_default.xml")
for i in tqdm(range(test.shape[0])):
    img = '/content/drive/My Drive/HAX/face_count/image_data/'+ test['Name'][i]
    img = cv2.imread(img ,0)
    face=facedetector.detectMultiScale(img,
                                 scaleFactor=1.1, 
                                 minNeighbors=3, 
                                 minSize=(20, 20)
                                 )
    faces.append(len(face))
```

```
print(faces)
```

```
submissions = pd.DataFrame({'Name':test['Name'],'HeadCount':faces})
submissions.to_csv("/content/drive/My Drive/HAX/face_count/submission.csv",index = False)
```
